```python
# this is spilliting the data into train and test data
 import numpy as np 
def spilt_data(data,test_ratio):
    indices=np.random.permutation(len(data))# generate the random suffeled index
    test_set_size=int(len(data)*test_ratio)
    test_indices=indices[:test_set_size]
    train_indice=indices[test_set_size:]
    return data.iloc[train_indices],data.iloc[test_indices]

```
